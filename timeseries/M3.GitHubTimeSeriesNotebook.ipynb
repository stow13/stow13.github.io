# Time Series

This is a randomly generated time series

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(5468)
N = 100
rng = pd.date_range('1903-01-01', freq='Y', periods=N)
df = pd.DataFrame(np.random.rand(N, 1) * 100, columns=['temp'], index=rng)
print(df)

fig, ax = plt.subplots()
fig.suptitle('Temperature Over Time', fontsize = 20)
fig.set_size_inches(15, 8)

plt.plot(df)
ax.set_ylim(0, 100)
ax.yaxis.set_label_text('Temperature', fontsize = 14) #X-axis caption
ax.xaxis.set_label_text('Time', fontsize = 14); #Y-axis caption
```
